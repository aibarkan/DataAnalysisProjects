```python
# Grouping the data by '<Parent ID>'
grouped = test_data.groupby('<Parent ID>')
values = []

for parent_id, group in grouped:
    consistent_cols = {}
    
    # Iterate over all columns to check for consistency within each group (except for '<ID>' and '<Parent ID>')
    for col in group.columns:
        if col not in ['<ID>', '<Parent ID>']:
            unique_values = group[col].dropna().unique()
            
            # If there's only one unique value and it exists in all rows, mark it as consistent
            if len(unique_values) == 1 and len(group[col].dropna()) == group.shape[0]:
                consistent_cols[col] = unique_values[0]  # Store the consistent value
    
    # Only create a value_row if there are consistent columns
    if consistent_cols:
        value_row = {'<Parent ID>': parent_id}  # Start with the Parent ID
        value_row.update(consistent_cols)  # Add the consistent columns to the row
        
        # Append the value_row to the final list of values
        values.append(value_row)

# Convert the final list of values into a DataFrame
if values:
    values_df = pd.DataFrame(values)
else:
    values_df = pd.DataFrame()

# Display the resulting DataFrame
values_df
```
