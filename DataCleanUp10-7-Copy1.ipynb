```python
import pandas as pd
```

```python
#input file for cleaning
data=pd.read_excel(r"")
export_file = r""
```

```python
data = pd.DataFrame(data)
```

```python
data.head()
```

```python
parent_sku_data = data[data['<Object Type Name>'].apply(lambda x: x in ['Product', 'SKU'])]
parent_sku_data
```

```python
grouped = data.groupby('<Parent ID>')
```

```python
results = []
values = []

for parent_id, group in grouped:
    non_blank_count = group.notna().sum().sum()
    average_non_blank = non_blank_count / len(group.columns)
    
    consistent_cols = {}
    

    for col in group.columns:
        unique_values = group[col].dropna().unique()
        if len(unique_values) == 1:
            consistent_cols[col] = unique_values[0]

    results.append({
        'parent_id': parent_id,
        'count_ids': group['<ID>'].nunique(),
        'avg_non_blank': average_non_blank,
        'non-unique_cols': consistent_cols
    })
    

    if consistent_cols:
        value_row = {'parent_id': parent_id}
        value_row.update(consistent_cols)  
        values.append(value_row)

result_df = pd.DataFrame(results)


if values:
    values_df = pd.DataFrame(values)
else:
    values_df = pd.DataFrame()

values_df
```

```python
with pd.ExcelWriter(export_file, mode='a') as writer:  values_df.to_excel(writer, sheet_name='TestOutputAll')
```
