## Heatmap

```python
#the lighter the color, higher correlation
plt.figure(figsize=(12,8))
sns.heatmap(df_excluded.corr())
plt.show()
```

```python
#the lighter the color, higher correlation
#each feature is positively correlated with itself
plt.figure(figsize=(12,8))
sns.heatmap(df_excluded.corr().abs())
plt.show()
```

# Stacked Histogram

```python
sns.histplot(df,x="mean radius", hue="target", multiple="stack")
```

```python
#if there is a lot of overlap between 2 bars, there is a predictable relationship
sns.histplot(df,x="worst concavity",hue="target",multiple="stack")
```

```python
columns = df.select_dtypes(exclude='object').drop("target",axis=1).columns
```

```python
columns
```

```python
#plot all variables using for loop to plot
fig , axs = plt.subplots(len(columns),1,figsize=(5,5*len(columns)))

for i in range(len(columns)):
    column_name = columns[i]
    sns.histplot(df,x=column_name,hue="target",multiple="stack",ax=axs[i])
```
