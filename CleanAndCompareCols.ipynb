```python
#Libraries Needed
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
export_file=r"filepath"
```

```python
#Dimension Attribute
dim_data = pd.read_excel(r"filepath")
col_compare = 'Col Name'
```

```python
#remove all 'X/x' leading/training space, str/""
```

```python
def clean_column(value):
    if pd.isna(value):  # Handle NaN values
        return value
    # Remove leading/trailing whitespaces and keep only numbers and decimal points
    return ''.join([char for char in str(value).strip() if char.isdigit()])

# Apply the cleaning function to the column
dim_data[col_compare] = dim_data[col_compare].apply(clean_column)
```

```python
unique_values = dim_data[col_compare].unique()
```

```python
unique_values
```

```python
dim_data["ConcateDim"] = dim_data[["Length", "Width", "Height", "Depth", "Multi-Dimension"]] \
    .astype(str) \
    .apply(lambda row: " ".join(row.replace("nan", "")), axis=1)
```

```python
unique_values = dim_data["ConcateDim"].unique()
```

```python
unique_values
```

```python
dim_data["ConcateDim"] = dim_data["ConcateDim"].apply(clean_column)
```

```python
# Safely compare the columns and include reverse check
dim_data['Comparison_Result'] = dim_data.apply(
    lambda row: "Match" if str(row['ConcateDim']) == str(row[col_compare]) 
                or str(row['ConcateDim']) == str(row[col_compare])[::-1] else "No Match",
    axis=1
)
```
